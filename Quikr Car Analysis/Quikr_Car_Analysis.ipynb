# Outliers Detection

```python
Categorical_data=[]
numerical_data=[]
for i in df:
    if df[i].dtype=='object':
        Categorical_data.append(i)
    else:
        numerical_data.append(i)
```

```python
print(Categorical_data)
print(numerical_data)
```

```python
for i in numerical_data:
    sns.histplot(df[i],kde=True)
    plt.show()
```

```python
for i in numerical_data:
    sns.boxplot(df[i])
    plt.show()
```

```python
df.describe()
```

```python
#Here i have used IQR method to remove outliers
for i in numerical_data:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    Lower_Whisker=Q1-1.5*IQR
    Upper_Whisker=Q3+1.5*IQR
    df=df[(df[i]>=Lower_Whisker) & (df[i]<=Upper_Whisker)]

#Price_Q1=np.percentile(df['Price'],25)
#Price_Q3=np.percentile(df['Price'],75)
#Price_IQR=Price_Q3-Price_Q1
#Price_Lower_Whisker=Price_Q1-1.5*Price_IQR
#Price_Upper_Whisker=Price_Q3+1.5*Price_IQR
```

# Hypothesis Testing

Conduct a hypothesis in above dataset that all cars are of 5 Lakhs or more than 5 Lakhs

take significant value= .05,no. of samples =30

H0 (Null Hypothesis) = All Cars value = 5 Lakhs

H1 (Altenate Hypothesis) = All Cars value > 5 Lakhs


```python
import scipy.stats as stats
```

```python
n=30
alpha=0.05
population_mean=500000
sample_mean=df['Price'].mean()
sample_std=df['Price'].std()
```

```python
tstat,p_value=stats.ttest_1samp(df['Price'],population_mean)
print("Stat value = "+str(tstat))
print("P value = "+str(p_value))
```

```python
if p_value < alpha:
    print("Reject the null hypothesis.\nValue of all cars are more than 5 Lakhs")
else:
    print("Accept the null hypothesis.\nValue of all the cars are 5 Lakhs")
```
